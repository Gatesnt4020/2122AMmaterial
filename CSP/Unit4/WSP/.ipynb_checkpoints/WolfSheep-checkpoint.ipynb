```python
import pandas as pd
import matplotlib.pyplot as plt
```

findStuff takes your filename and grass is if your setting has grass then its True

```python
def findStuff(FILENAME,grass):
    df = pd.read_csv(FILENAME,header=0)
    sheep=df['sheepY']
    wolf=df['wolfY']
    sheepMax=sheep.max()
    sheepMin=sheep.min()
    sheepAve=sheep.mean()
    print("sheep max ",sheepMax," sheep min ",sheepMin," sheep ave ",round(sheepAve))
    wolfMax=wolf.max()
    wolfMin=wolf.min()
    wolfAve=wolf.mean()
    print("wolf max ",wolfMax," wolf min ",wolfMin," wolf ave ",round(wolfAve))
    if grass:
        g=df['grassY']
        gMax=g.max()
        gMin=g.min()
        gAve=g.mean()
        print("grsas max ",gMax," grass min ",gMin," grass ave ",round(gAve))
    reoccur=0
    for i in range(len(sheep)):
        if grass:
            if sheep[i] == wolf[i] or sheep[i] == g[i] or wolf[i] == g[i]:
                reoccur+=1
            elif sheep[i] == wolf[i]:
                reoccur+=1
    print("times were the species approximately the same ",reoccur)
```

```python
def makeGraphs(FILENAME,titly,xlab,ylab,grass):
    df = pd.read_csv(FILENAME,header=0)
    sY=df['sheepY']
    wY=df['wolfY']
    sX=df['sheepX']
    wX=df['wolfX']
    plt.plot(sX,sY,color='tan',label="sheep")
    plt.plot(wX,wY,color='black',label="wolf")
    if grass:
        gY=df['grassY']
        gX=df['grassX']
        plt.plot(gX,gY,color='green',label="grass")
    
    
    plt.title(titly)
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.legend()
    plt.show()
```

### Model Settings
- wolf-gain-from-food = 20
- show-energy? = FALSE
- initial-number-wolves = 50
- wolf-reproduce = 5
- initial-number-sheep = 100
- model-version = "sheep-wolves"
- sheep-gain-from-food = 4
- grass-regrowth-time = 30
- sheep-reproduce = 4

```python
findStuff("sheepWin.csv",False)
```

```python
makeGraphs("sheepWin.csv","Sheep Win","Time","Population",False)
```
