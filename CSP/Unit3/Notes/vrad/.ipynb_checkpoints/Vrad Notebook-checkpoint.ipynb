# Import file and ecessary libraries 
The file we are importing is a class that we created to reformat our vins. 
In a vin, there are 3 main sections
- wmi = first 3
- vds = second 6
- ser = remaining items

"vins.csv" is our original data

```python
from VIN import VIN
with open("vins.csv","r") as f:
    file = f.readlines()
    for i in range(len(file)):
        file[i] = VIN(file[i].rstrip())
```

Explain more about my function below

```python
#this function will convert the letters to a base 10 number
def vmapenc(vin):
    vmap="0123456789ABCDEFGHJKLMNPRSTUVWXYZ"
    vrad = 0  #vrad will be out output
    exp = 0  #exponent number
    base = 34  #34 possible characters
    for v in range(len(vin)):
        coe = vmap.index(vin[v])
        exp = len(vin)-1-v
        vrad+= coe*base**exp
    return vrad
```

```python
#loop through your file and convert each item in the file
for v in file:
    print(vmapenc(v.wmi+v.vds+v.ser))
```

If you notice from the run above, those numbers are <h3>MASSIVE</h3>
Break down each number into 3 section, the wmi, vds, and ser

```python
for v in file:
    print(vmapenc(v.wmi))
    print(vmapenc(v.vds))
    print(vmapenc(v.ser))
```
